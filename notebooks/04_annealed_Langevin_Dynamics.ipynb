```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 10  # Number of noise levels
T = 1000  # Number of time steps for each noise level
sigma_1 = 100  # Initial sigma (noise level)
sigma_L = 0.01  # Final sigma
epsilon = 1e-5  # Small value to control the learning rate

# Define noise levels, from sigma_1 (large) to sigma_L (small)
sigma = np.logspace(np.log10(sigma_1), np.log10(sigma_L), L)

# Define step size alpha_i
alpha = epsilon * np.square(sigma) / np.square(sigma_L)

# Plot the alpha values for each sigma layer
plt.figure(figsize=(8, 6))
plt.plot(sigma, alpha, marker='o', label="Alpha (step size)")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Sigma (Noise Level)")
plt.ylabel("Alpha (Step Size)")
plt.title("Alpha Values for Each Sigma Level in Annealed Langevin Dynamics")
plt.grid(True)
plt.legend()
plt.show()
```
